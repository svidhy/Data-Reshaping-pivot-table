#### We want to know the information about the mean salary and number of employees in each department and for each job title.

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
employee.groupby(['Department','Title']).agg({'Salary':['count', np.median ,min, max, np.std]}).reset_index()
```

```python
employee.pivot_table(index=['Department', 'Title'], values=['Salary'], aggfunc = ['count', np.median ,min, max, np.std]).reset_index()
```

### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
# your answer here
#mean_salary will find the index of the max and then we drop it from the list. After dropping, we return the mean.

def mean_salary(x):
    ind = x.idxmax(axis=0)
    #print(ind)
    x = x.drop(ind)
    return x.mean()


employee.groupby(['Department']).agg({'Salary': mean_salary}).reset_index()
```
